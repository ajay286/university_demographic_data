```python
import pandas as pd
df = pd.read_csv('Data_source\\enrolment_unique_headcount_0.csv')
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(15,7))
ax.set_xlabel("Fiscal year")
ax.set_ylabel("Unique Headcount")
df.groupby(['Fiscal Year','Gender']).count()['Unique Headcount'].unstack().plot(ax=ax,title='Different genders Headcount over the years')
```

```python
df.groupby(['Program Grouping'])['Unique Headcount'].mean().sort_values().tail(20).plot(kind='barh',figsize=(20,20),title='Top 20 programs with Headcount')
```

```python
campus_groups = df.groupby('Campus').count()['Unique Headcount']
campus_groups.plot.pie(figsize=(6,6),   autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2, title='Percentage per campus')
```

```python
# Program grouping.
df.groupby(['Program Grouping'])['Unique Headcount'].mean().plot.bar(figsize=(50,50))
```
